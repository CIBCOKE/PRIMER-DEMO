###Bifurcaciones: if ... else ...
La estructura de control if ... else ... permite que un programa ejecute unas instrucciones cuando se cumple una condicin y otras instrucciones cuando no se cumple esa condicin. En ingls "if" significa "si" (condicin) y "else" significa "si no". La orden en Python se escribe as:

###Sintaxis de la sentencia condicional if ... else ...
La sintaxis de la construccin if ... else ... es la siguiente:
```
if condicin:
    aqu van las rdenes que se ejecutan si la condicin es cierta
    y que pueden ocupar varias lneas
else:
    y aqu van las rdenes que se ejecutan si la condicin es
    falsa y que tambin pueden ocupar varias lneas
```

La ejecucin de esta construccin es la siguiente:
* La condicin se evala siempre.
** Si el resultado es True se ejecuta solamente el bloque de sentencias 1
** Si el resultado es False se ejecuta solamente el bloque de sentencias 2.

La primera lnea contiene la condicin a evaluar. Esta lnea debe terminar siempre por dos puntos (:).

A continuacin viene el bloque de rdenes que se ejecutan cuando la condicin se cumple (es decir, cuando la condicin es verdadera). Es importante sealar que este bloque debe ir sangrado, puesto que Python utiliza el sangrado para reconocer las lneas que forman un bloque de instrucciones. El sangrado que se suele utilizar en Python es de cuatro espacios, pero se pueden utilizar ms o menos espacios. Al escribir dos puntos (:) al final de una lnea, COLAB sangrar automticamente las lneas siguientes. Para terminar un bloque, basta con volver al principio de la lnea.

Despus viene la lnea con la orden else, que indica a Python que el bloque que viene a continuacin se tiene que ejecutar cuando la condicin no se cumpla (es decir, cuando sea falsa). Esta lnea tambin debe terminar siempre por dos puntos (:). La lnea con la orden else no debe incluir nada ms que el else y los dos puntos.

En ltimo lugar est el bloque de instrucciones sangrado que corresponde al else.

Es mejor no hacerlo as por dos motivos:
* al poner dos bloques if estamos obligando a Python a evaluar siempre las dos condiciones, mientras que en un bloque if ... else ... slo se evala una condicin. En un programa sencillo la diferencia no es apreciable, pero en programas que ejecutan muchas comparaciones, el impacto puede ser apreciable.
* utilizando else nos ahorramos escribir una condicin (adems, escribiendo la condicin nos podemos equivocar, pero escribiendo else no).
* utilizando if ... else nos aseguramos de que se ejecuta uno de los dos bloques de instrucciones. Utilizando dos if cabra la posibilidad de que no se cumpliera ninguna de las dos condiciones y no se ejecutara ninguno de los dos bloques de instrucciones.

###Ms de dos alternativas: if ... elif ... else ...

La construccin if ... else ... se puede extender aadiendo la instruccin elif:

La estructura de control if ... elif ... else ... permite encadenar varias condiciones. elif es una contraccin de else if. La orden en Python se escribe as:

####Sintaxis de la sentencia condicional if ... elif ... else ...
La sintaxis de la construccin if ... elif ... else ... es la siguiente:
```
if condicin_1:
    bloque 1
elif condicin_2:
    bloque 2
else:
    bloque 3
```

* Si se cumple la condicin 1, se ejecuta el bloque 1
* Si no se cumple la condicin 1 pero s que se cumple la condicin 2, se ejecuta el bloque 2
* Si no se cumplen ni la condicin 1 ni la condicin 2, se ejecuta el bloque 3.

Esta estructura es equivalente a la siguiente estructura de if ... else ... anidados:
```
if condicin_1:
    bloque 1
else:
    if condicin_2:
        bloque 2
    else:
        bloque 3
```

Se pueden escribir tantos bloques elif como sean necesarios. El bloque else (que es opcional) se ejecuta si no se cumple ninguna de las condiciones anteriores.

###Bucle While
En Python tiene una palabra reservada llamada while que nos permite ejecutar ciclos, o bien secuencias peridicas que nos permiten ejecutar cdigo mltiples veces.
El ciclo while nos permite realizar mltiples iteraciones basndonos en el resultado de una expresin lgica que puede tener como resultado un valor True o False.

```
     #Bucle while
```

####El ejemplo sigiente tiene un contador con un valor inicial de cero, cada iteracin del while manipula esta variable de manera que incremente su valor en 1, por lo que despus de su primera iteracin el contador tendr un valor de 1, luego 2, y as sucesivamente.

Eventualmente cuando el contador llegue a tener un valor de 10, la condicin del ciclo numero <= 10 sera False, por lo que el ciclo terminar arrojando el siguiente resultado.

````
#Celda 30
suma, numero = 0, 1

while numero <= 10:
    suma = numero + suma
    numero = numero + 1
    
print("La suma es " ,suma)
````

####En el ejemplo siguiente el evento que se dispara cuando el usuario ingresa el valor -1, causando que el bucle while termine.

````
#Celda 31
promedio, total, contar = 0.0, 0, 0

grado = int(input("Introduzca la nota de un estudiante (-1 para salir): "))	
while grado != -1:
    total = total + grado
    contar = contar + 1
    grado = int(input("Introduzca la nota de un estudiante (-1 para salir): "))
promedio = total / contar
print("Promedio de notas del grado escolar es: ",promedio)
````

###Bucle while con else.-
####Al igual que la sentencia if, la estructura while tambin puede combinarse con una sentencia else).

El nombre de la sentencia else es equivocada, ya que el bloque else se ejecutar en todos los casos, es decir, cuando la expresin condicional del while sea False, (a comparacin de la sentencia if).

````
#Celda 32
promedio, total, contar = 0.0, 0, 0
mensaje = "Introduzca la nota de un estudiante (-1 para salir): "

grado = int(input(mensaje))
while grado != -1:
    total = total + grado
    contar += 1
    grado = int(input(mensaje))
else:
    promedio = total / contar
    print("Promedio de notas del grado escolar: ",promedio)
````

###Sentencia break.-
####A continuacin, se presenta un ejemplo del uso del bucle while controlado la sentencia break:

````
#Celda 33
variable = 10

while variable > 0:
    print('Actual valor de variable:', variable)
    variable = variable -1
    if variable == 5:
        break
````

Adicionalmente existe una forma alternativa de interrumpir o cortar los ciclos utilizando la palabra reservada break.

Esta nos permite salir del ciclo incluso si la expresin evaluada en while (o en otro ciclo como for) permanece siendo True.

###Sentencia continue.-
####A continuacin, se presenta un ejemplo del uso del bucle while controlado la sentencia continue:

````
#Celda 34
variable = 10

while variable > 0:              
   variable = variable -1
   if variable == 5:
      continue
   print('Actual valor de variable:', variable)
````

La sentencia continue hace que pase de nuevo al principio del bucle aunque no se haya terminado de ejecutar el ciclo anterior.
